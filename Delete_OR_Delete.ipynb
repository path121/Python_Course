# Delete / Removing:
- It used to removing the element from array

- Using Delete Function

### Synntax:
- **np.delete(array, index_number, axis = None)**

```python
import numpy as np

arr = np.array([10,20,30,40,50,60])

new_arr = np.delete(arr, 1, axis=0)
```

```python
print('First Array:')
print(arr)

print('\nNew array:')
print(new_arr)
```

**20 Is Deleted From Array**

# Delete / Removing In Multi-Dimensional Array
- It means direct deleting row / column
- **axis = 0** : It means deleting row.
- **axis = 1** : It means deleting column.

```python
arr_2d = np.array([[1,2,3],[4,5,6]])

new_arr_2d = np.delete(arr_2d, 0, axis = 0)
```

```python
print(new_arr_2d)
```
