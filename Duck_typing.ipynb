```python
# Q1 — Easy
# Create 3 classes Dog, Cat, Bird
# • Each has a method speak()
# • Dog → Woof!, Cat → Meow!, Bird → Tweet!
# • Create a function make_sound(animal)
# • Call it with all 3 objects in a loop

class Dog:
    def speak(self):
        print("Woof!")


class Cat:
    def speak(self):
        print("Meow!")


class Bird:
    def speak(self):
        print("Tweet!")


def make_sound(animal):
    animal.speak()


animals = [Dog(), Cat(), Bird()]

for animal in animals:
    make_sound(animal)
```

```python
# Q2 — Easy
# Create 3 classes Laptop, Mobile, Tablet
# • Each has a method turn_on() and turn_off()
# • Laptop → Windows starting!, Mobile → Android starting!
# • Tablet → iPad starting!
# • Create function start_device(device)
# • Call turn_on() and turn_off() for all devices

class Laptop:
    def turn_on(self):
        print("Windows starting!")

    def turn_off(self):
        print("Shutting Down")

class Mobile:
    def turn_on(self):
        print("Android starting!")

    def turn_off(self):
        print("shutting Down")

class Tablet:
    def turn_on(self):
        print("iPad starting!")

    def turn_off(self):
        print("Shutting Down")


def start_device(device):
    device.turn_on()
    device.turn_off()

devices  = [Laptop(), Mobile(), Tablet()]

for device in devices:
    start_device(device)
    
```

```python
# Q3 — Easy
# Create 3 classes Chef, Doctor, Teacher
# • Each has a method work() and introduce()
# • Chef → cooking food, Doctor → treating patients
# • Teacher → teaching students
# • Create function do_work(person)
# • Loop through all 3 objects and call both methods

class Chef:
    def work(self):
        print("cooking food")

    def introduce(self):
        print("Chef")

class Doctor:
    def work(self):
        print("treating patients")

    def introduce(self):
        print("Doctor")

class Teacher:
    def work(self):
        print("teaching students")

    def introduce(self):
        print("Teacher")


def do_work(person):
    person.work()
    person.introduce()

persons = [Chef(), Doctor(), Teacher()]

for person in persons:
    do_work(person)
    print()
```

```python
# Q4 — Medium
# Create 4 classes YouTube, Netflix, Spotify, Podcast
# • Each has methods play(title) and pause() and stop()
# • Create function stream(platform, title)
# • Each platform shows its OWN message when playing


# • Loop and test all platforms with different titles

class YouTube:
    def play(self, title):
        print(f"YouTube is playing: {title}")

    def pause(self):
        print("YouTube paused")

    def stop(self):
        print("YouTube stopped")


class Netflix:
    def play(self, title):
        print(f"Netflix is streaming: {title}")

    def pause(self):
        print("Netflix paused")

    def stop(self):
        print("Netflix stopped")


class Spotify:
    def play(self, title):
        print(f"Spotify is playing music: {title}")

    def pause(self):
        print("Spotify paused")

    def stop(self):
        print("Spotify stopped")


class Podcast:
    def play(self, title):
        print(f"Podcast is playing: {title}")

    def pause(self):
        print("Podcast paused")

    def stop(self):
        print("Podcast stopped")


def stream(platform, title):
    platform.play(title)
    platform.pause()
    platform.stop()


platforms = [
    (YouTube(), "Python Tutorial"),
    (Netflix(), "Stranger Things"),
    (Spotify(), "Believer"),
    (Podcast(), "Data Science Talk")
]

for platform, title in platforms:
    stream(platform, title)
    print("*")
```

```python
# Q5 — Medium
# Create 4 classes PDF, Word, Excel, PowerPoint
# • Each has methods:
# o open(filename) → shows opening message
# o save(filename) → shows saving message
# o close() → shows closing message
# • Create function handle_file(app, filename)
# • Test all 4 file types

class PDF:
    def open(self, filename):
        print(f"opening {filename} file")
    def save(self, filename):
        print(f"saving {filename} file")
    def close(self):
        print("Closing PDF File")

    
class Word:
    def open(self, filename):
        print(f"opening {filename} file")
    def save(self, filename):
        print(f"saving {filename} file")
    def close(self):
        print("Closing Word File")

class Excel:
    def open(self, filename):
        print(f"opening {filename} file")
    def save(self, filename):
        print(f"saving {filename} file")
    def close(self):
        print("Closing Excel File")


class PowerPoint:
    def open(self, filename):
        print(f"opening {filename} file")
    def save(self, filename):
        print(f"saving {filename} file")
    def close(self):
        print("Closing PowerPoint File")

        
def handle_file(app, filename):
    app.open(filename)
    app.save(filename)

apps = [
    (PDF(),"pdf"),
    (Word(),"Word"),
    (Excel(),"Excel"),
    (PowerPoint(),"PowerPoint")

]

for app, filename in apps:
    handle_file(app, filename)
    print("~" *55)
    
```

```python
# Q6 — Medium
# Create 4 classes for delivery:
# BikeDelivery, CarDelivery, DroneDelivery, BoatDelivery
# • Each has methods:
# o pickup(item) → picks up item
# o deliver(item, address) → delivers to address
# o confirm() → confirms delivery
# • Create function process_delivery(vehicle, item, address)
# • Test all 4 delivery types

class BikeDelivery:
    def pickup(self, item):
        print(f"Your {item} picked up by Bike")
    def deliver(self, item, address):
        print(f"your {item} address is {address}")
    def confirm(self):
        print("Bike delivery confirmed")

class CarDelivery:
    def pickup(self, item):
        print(f"Your {item} picked up by Car")
    def deliver(self, item, address):
        print(f"your {item} address is {address}")
    def confirm(self):
        print("Car delivery is confirmed")

class DroneDelivery:
    def pickup(self, item):
        print(f"Your {item} picked up by Drone")
    def deliver(self, item, address):
        print(f"your {item} address is {address}")
    def confirm(self):
        print("Drone delivery is confirm")

class BoatDelivery:
    def pickup(self, item):
        print(f"Your {item} picked up by Boat")
    def deliver(self, item, address):
        print(f"Your {item} delivered to {address} by Bike")
    def confirm(self):
        print("Boat delivery confirmed")

def process_delivery(vehicle, item, address):
    vehicle.pickup(item)
    vehicle.deliver(item, address)
    vehicle.confirm()

vehicles = [
    BikeDelivery(),
    CarDelivery(),
    DroneDelivery(),
    BoatDelivery()
    

]

for vehicle in vehicles:
    process_delivery(vehicle, "Phone","Tirthpuri" )
    print()
```

```python
# Q7 — Medium
# Create 4 classes for payments:
# CreditCard, DebitCard, UPI, NetBanking
# • Each has:
# o private __balance
# o method validate(amount) → checks if balance enough
# o method pay(amount) → deducts and shows message
# o method receipt() → shows receipt
# • Create function make_payment(method, amount)

class CreditCard:
    def __init__(self, balance):
        self.__balance = balance

    def validate(self, amount):
        return self.__balance >= amount

    def pay(self, amount):
        if self.validate(amount):
            self.__balance -= amount
            print(f"CreditCard Payment of {amount} Successful")
        else:
            print("Insufficient Balance")

    def receipt(self):
        print("Receipt Generated")


class DebitCard:
    def __init__(self, balance):
        self.__balance = balance

    def validate(self, amount):
        return self.__balance >= amount

    def pay(self, amount):
        if self.validate(amount):
            self.__balance -= amount
            print(f"DebitCard Payment of {amount} Successful")
        else:
            print("Insufficient Balance")

    def receipt(self):
        print("Receipt Generated")


class UPI:
    def __init__(self, balance):
        self.__balance = balance

    def validate(self, amount):
        return self.__balance >= amount

    def pay(self, amount):
        if self.validate(amount):
            self.__balance -= amount
            print(f"UPI Payment of {amount} Successful")
        else:
            print("Insufficient Balance")

    def receipt(self):
        print("Receipt Generated")


class NetBanking:
    def __init__(self, balance):
        self.__balance = balance

    def validate(self, amount):
        return self.__balance >= amount

    def pay(self, amount):
        if self.validate(amount):
            self.__balance -= amount
            print(f"NetBanking Payment of {amount} Successful")
        else:
            print("Insufficient Balance")

    def receipt(self):
        print("Receipt Generated")


def make_payment(method, amount):
    method.pay(amount)
    print("~" *55)
    method.receipt()


c = CreditCard(10000)
d = DebitCard(5000)
u = UPI(3000)
n = NetBanking(8000)

make_payment(c, 5000)
make_payment(d, 1000)
make_payment(u, 500)
make_payment(n, 3000)
```
