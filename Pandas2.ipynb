### #*MEDIUM LEVEL (Business Logic + Grouping)*


```python
# 1. Find department-wise average salary.


df.groupby("Department")["Salary"].mean().reset_index(name="Average_Salary")
```

```python
# 2. Which department has the highest average performance rating?

df.groupby("Department")["Performance_Rating"].mean().sort_values(ascending=False).head(1)
```

```python
# 3. Identify employees whose salary is above company average.

df.loc[df["Salary"]> df["Salary"].mean(),["Name","Salary"]]
```

```python
# 4. Create an Age Group column:

#    20–30 → Young

#    31–45 → Mid

#    46+ → Senior
#    Then count employees per age group.

import numpy as np

df["Age_Group"] = np.where(
    (df["Age"] >= 20) & (df["Age"] <= 30), "Young",
    np.where(
        (df["Age"] >= 31) & (df["Age"] <= 45), "Mid",
        "Senior"
    )
)


df["Age_Group"].value_counts()
```

```python
def age_group(age):
    if 20 <= age <= 30:
        return "Young"
    elif 31 <= age <= 45:
        return "Mid"
    else:
        return "Senior"

df["Age_Group"] = df["Age"].apply(age_group)
```

```python
df.value_counts("Age_Group")
```

```python
# 5. Find employees whose experience is greater than department average experience.

df.loc[ df["Experience_Years"] > df.groupby("Department")["Experience_Years"].transform("mean"), ["Name", "Department", "Experience_Years"] ]
```

```python
df.loc[df["Experience_Years"]> df["Experience_Years"].mean(), ["Name", "Department", "Experience_Years"]]
```

```python
# 6. Calculate salary growth ratio:
#    Salary / Experience\_Years
#    Find top 10 employees with highest ratio.

df["Salary_Growth_Ratio"] = df["Salary"] / df["Experience_Years"]
```
