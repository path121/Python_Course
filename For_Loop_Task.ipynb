```python
# 1. Print Names Starting With "A"
# names = ["Aman", "Ravi", "Ankit", "Sumit", "Ajay"]
# Print only names starting with "A".

names = ["Aman", "Ravi", "Ankit", "Sumit", "Ajay"]
for i in names:
    if i.lower().startswith("a"):
        print(i)


```

```python
# 2. Print Numbers Ending With 5
# nums = [12, 25, 30, 45, 67, 95]
# Print numbers whose last digit is 5.

nums = [12, 25, 30, 45, 67, 95]
for i in nums:
    if str(i).endswith("5"):
        print(i)
```

```python
# 3. Count Names Containing "a"
# names = ["Rahul", "Aman", "Riya", "Karan", "Sonia"]
# Count how many names contain letter "a".

names = ["Rahul", "Aman", "Riya", "Karan", "Sonia"]
for i in names:
    if i.lower().count("a"):
        print(i)
```

```python
# 4. Print Word With Length
# words = ["Python", "Java", "SQL", "Pandas"]
# Print each word and its length.

words = ["Python", "Java", "SQL", "Pandas"]
for i in words:
    if len(i):
        print(f"{i} : {len(i)}")
```

```python
# 5. Divisible by Both 2 and 3
# nums = [6, 8, 12, 15, 18, 20, 24]
# Print numbers divisible by both 2 and 3.

nums = [6, 8, 12, 15, 18, 20, 24]
for i in nums:
    if i % 2==0 and i % 3==0:
        print(f"divisible by 2 And 3 : {i}")
```

```python
# 6. Reverse Every String
# words = ["cat", "dog", "apple", "python"]
# Reverse every string using a for loop.

words = ["cat", "dog", "apple", "python"]
for i in words:
    print(i [::-1])
```

```python
# 7. Find Smallest Number
# nums = [45, 12, 89, 3, 67, 21]
# Find the smallest number using a loop.

nums = [45, 12, 89, 3, 67, 21]

smallest = nums[0]

for i in nums:
    if i < smallest:
        smallest = i

print("Smallest number:", smallest)
```

```python
# 8. Print Vowels From Words
# words = ["apple", "sky", "orange", "fly"]
# Print only vowels from each word.

words = ["apple", "sky", "orange", "fly","DHANRaJ"]

for i in words:
    print("Word:", i)
    for ch in i:
        if ch in "aeiouAEIOU":
            print(ch)
```

```python
# 9. Square of Odd Numbers
# nums = [1, 2, 3, 4, 5, 6, 7]
# Create a new list containing square of odd numbers.

nums = [1, 2, 3, 4, 5, 6, 7]
for i in nums:
    if i % 2 !=0:
        print(f"{i} : {i ** 2}")
```

```python
# 10. Print Duplicate Values
# nums = [1, 2, 2, 3, 4, 4, 5, 1]
# Print duplicate values only.

nums = [1, 2, 2, 3, 4, 4, 5, 1]
for i in nums:
    if nums.count(i) > 1:
        print(i)
```

```python
# 11. Count Even and Odd Numbers
# nums = [11, 22, 33, 44, 55, 66]
# Count how many even and odd numbers are present.

nums = [11, 22, 33, 44, 55, 66,1,3,3]

count = 0

for i in nums:
    if i % 2 != 0:
        count += 1

print("odd numbers:", count)
```

```python
# 12. Convert First Letter to Uppercase
# names = ["sumit", "aman", "ravi", "karan"]
# Convert first letter of every word to uppercase.

names = ["sumit", "aman", "ravi", "karan"]
for i in names:
    print(i.upper())
```
