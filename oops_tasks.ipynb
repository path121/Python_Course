```python
# Encapsulation
# 3. Bank Account System
# Create a class BankAccount.
# Private Attribute:
# • __Balance

# Methods:
# • Deposit()
# • Withdraw()
# • Show_Balance()
# Concept
# Hide balance from direct access and update it through methods.

class BankAccount:
    def __init__(self, __Balance):
        self.__Balance =  __Balance

    def deposit(self, deposit):
        self.__Balance += deposit
        print("Balance Successfully deposit")

    def withdraw(self, withdraw):
        self.__Balance -= withdraw
        print("Balance Successfully withdraw")

    def Show_Balance(self):
        print("Balance: ",self.__Balance)

b1 = BankAccount(5000)
b1.Show_Balance()
```

```python
b1.withdraw(500)

b1.Show_Balance()
```

```python
# 6. ATM System
# Create a class ATM.

# Private Attribute:
# • __Balance
# Methods:
# • Deposit()
# • Withdraw()
# Condition:
# • Do not allow withdrawal if balance is insufficient.
# Concept
# Real-world encapsulation example.

class ATM:
    def __init__(self, balance):
        self.__balance = balance

    def Deposit(self, deposit):
        self.__balance += deposit
        print(f"Balance Successfully deposit ₹{deposit}")

    def withdraw(self,withdraw):
        if withdraw > self.__balance:
            print("balance is insufficient!!...")

        else :
            
            self.__balance -= withdraw
            print(f"Balance Successfully withdraw ₹")

    def show_balance(self):
        print("Balance: ",self.__balance)


c1 = ATM(5000)
c1.show_balance()
```

```python
# 7. User Password System
# Create a class User.
# Private Attribute:
# • __Password
# Methods:
# • Change_Password()
# • Validate_Password()
# Concept
# Hide sensitive information.

class User:

    def __init__(self, Password):
        
        self.__Password = Password

    def Change_Password(self, new_pass):
        self.__Password = new_pass
        print("Password Changed Successfully")


        
    def validate_password(self, Password):
        if self.__Password == Password:
            print("Valid Password")
        else:
            print("Invalid Password")

    def display_pass(self):
        print("Password: ",self.__Password)

u1 = User("admin123")
u1.display_pass()
print("*" * 55)

u1.validate_password("admin123")

u1.Change_Password("Parth@3737")
u1.validate_password("Parth@3737")
print("*" * 55)

u1.display_pass()
```

```python
# Single Inheritance
# 8. Employee → Developer
# Parent
# Employee
# Attributes:
# • Employee_Name
# • Employee_ID
# Child

# Developer
# Attribute:
# • Programming_Language
# Concept
# Use super() to inherit employee details.

class Employee:
    def __init__(self, Employee_Name, Employee_ID):
        self.Employee_Name = Employee_Name
        self.Employee_ID = Employee_ID

class Developer(Employee):
    def __init__(self, Employee_Name, Employee_ID, Programming_Language):
        super().__init__(Employee_Name, Employee_ID)
        self.Programming_Language = Programming_Language

    def display(self):
        print("Employee_Name: ",self.Employee_Name)
        print("Employee_ID: ",self.Employee_ID)
        print("Programming_Language: ",self.Programming_Language)

e1 = Developer("Dhanraj", "Abc457895", "Python")
e1.display()
```

```python
# 10. School → Student
# Parent
# School
# Attributes:
# • School_Name
# • City
# Child
# Student
# Attribute:

# • Student_Name
# Concept
# Reuse parent attributes using super().

class School:
    def __init__(self, school_name, city):
        self.school_name = school_name
        self.city = city

class Student(School):
    def __init__(self, school_name, city, student_name):
        super().__init__(school_name, city)
        self.student_name = student_name

    def display(self):
        print("school_name: ",self.school_name)
        print("city: ",self.city)
        print("student_name: ",self.student_name)

s1 = Student("MSS", "Tirthpuri", "Dhanraj")
s1.display()
```

```python
# Multilevel Inheritance
# 11. Person → Employee → Manager
# Person
# • Name
# Employee
# • Employee_ID
# Manager
# • Department
# Concept
# Learn constructor chaining through multiple levels.

class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):
    def __init__(self, name,employee_id):
        super().__init__(name)
        self.employee_id = employee_id

class Manager(Employee):
    def __init__(self, name, employee_id, department):
        super().__init__(name,employee_id)
        self.department = department

    def display(self):
        print("name: ",self.name)
        print("employee_id: ",self.employee_id)
        print("department: ",self.department)

p1 = Manager("Dhanraj", "Abc54789650", "Data Analyst")
p1.display()
```

```python
# 12. Company → Department → Employee
# Company
# • Company_Name
# Department
# • Department_Name
# Employee
# • Employee_Name
# Concept
# Pass data through three inheritance levels.

class Company:
    def __init__(self, company_name):
        self.company_name = company_name

class Department(Company):
    def __init__(self, company_name, Department_Name):
        super().__init__(company_name)
        self.Department_Name = Department_Name

class Employee(Department):
    def __init__(self, company_name, Department_Name, Employee_Name):
        super().__init__(company_name, Department_Name)
        self.Employee_Name = Employee_Name

    def display_company(self):
        print("company_name: ", self.company_name)
        print("Department_Name: ", self.Department_Name)
        print("Employee_Name: ", self.Employee_Name)

c1 = Employee("IBM", "Data Science", "Dhanraj")
c1.display_company()
```

```python
# 13. University → College → Student
# University
# • University_Name

# College
# • College_Name

# Student
# • Student_Name

# Concept
# Practice multilevel inheritance with super().

class University:
    def __init__(self,University_Name):
        self.University_Name = University_Name

class College(University):
    def __init__(self,University_Name, College_Name):
        super().__init__(University_Name)
        self.College_Name = College_Name
        
class Student(College):
    def __init__(self,University_Name, College_Name, Student):
        super().__init__(University_Name, College_Name)
        self.Student = Student

    def display_university(self):
        print("University_Name: ",self.University_Name)
        print("College_Name: ",self.College_Name)
        print("Student: ",self.Student)

u1 = Student("BAMU", "Deogiri College", "Markad Dhanraj")
u1.display_university()
```

```python
# Multiple Inheritance
# 14. Father + Mother → Child
# Father
# • Father_Name
# • Father_Job
# Mother
# • Mother_Name
# • Mother_Job
# Child
# • Child_Name
# Concept
# Use multiple inheritance and display all family details.

class Father:
    def __init__(self, father_name, father_job):
        self.father_name = father_name
        self.father_job = father_job

class Mother:
    def __init__(self, mother_name, mother_job):
        self.mother_name = mother_name
        self.mother_job = mother_job

class Child(Father,Mother):
    def __init__(self, father_name, father_job, mother_name, mother_job, child_name):
        Father.__init__(self, father_name, father_job)
        Mother.__init__(self, mother_name, mother_job)
        self.child_name = child_name

    def display_family_details(self):
        print("Father_Name: ", self.father_name)
        print("Father_Job: ", self.father_job)
        print("Mother_Name: ", self.mother_name)
        print("Mother_Job: ", self.mother_job)
        print("Child_Name: ", self.child_name)

f1 = Child("Appasaheb", " Farmer", "Ratnamala", " House_Wife", " Dhanraj & Pruthviraj")
f1.display_family_details()
```

```python
# Hybrid Inheritance
# 17. University → College + School → Student
# University
# • University_Name
# • City
# College
# • College_Name
# School
# • School_Name
# Student
# • Student_Name
# Concept
# Combination of Hierarchical + Multiple Inheritance.

class University:
    def __init__(self, University_Name, city):
        self.University_Name = University_Name
        self.city = city


class College(University):
    def __init__(self,  University_Name, city, College_Name):
        University.__init__(self, University_Name, city)
        self.College_Name = College_Name


class School(University):
    def __init__(self, University_Name, city, School_Name):
        University.__init__(self, University_Name, city)
        self.School_Name = School_Name

class Student(College, School):
    def __init__(self, University_Name, city, College_Name, School_Name, Student_Name):
        College.__init__(self, University_Name, city, College_Name)
        School.__init__(self, University_Name, city, School_Name)
        self.Student_Name = Student_Name

    def display(self):
        print("University_Name: ",self.University_Name)
        print("city: ",self.city)
        print("College_Name: ",self.College_Name)
        print("School_Name: ",self.School_Name)
        print("Student_Name: ",self.Student_Name)
        
u1 = Student("BAMU", "Chh. Sambhajinagar", "Deogiri College", "Mss", "Dhanraj")
u1.display() 
```

```python
# 18. Hospital → Doctor + Nurse → Patient
# Hospital

# • Hospital_Name
# Doctor
# • Doctor_Name
# Nurse
# • Nurse_Name
# Patient
# • Patient_Name
# Concept
# Hybrid inheritance with shared parent.

class Hospital:
    def __init__(self, Hospital_Name):
        self.Hospital_Name = Hospital_Name

class Doctor(Hospital):
    def __init__(self, Hospital_Name, Doctor_Name):
        Hospital.__init__(self, Hospital_Name)
        self.Doctor_Name = Doctor_Name

class Nurse(Hospital):
    def __init__(self, Hospital_Name, Nurse_Name):
        Hospital.__init__(self, Hospital_Name)
        self.Nurse_Name = Nurse_Name

class Patient(Doctor, Nurse):
    def __init__(self, Hospital_Name, Doctor_Name, Nurse_Name, Patient_Name):
        Doctor.__init__(self, Hospital_Name, Doctor_Name)
        Nurse.__init__(self, Hospital_Name, Nurse_Name)
        self.Patient_Name = Patient_Name


    def display(self):
        print("Hospital_Name: ",self.Hospital_Name)
        print("Doctor_Name: ",self.Doctor_Name)
        print("Nurse_Name: ",self.Nurse_Name)
        print("Patient_Name: ",self.Patient_Name)

h1 = Patient("Vighnaharta Hospital", "Dr.Pruthviraj", "Sunita", "Shaila")
h1.display()
```

```python
# Abstraction
# 19. Bank System

# Abstract Class
# Bank
# Attributes:
# • Bank_Name
# • Branch

# Abstract Method:
# • Show_Interest_Rate()

# Child Class
# SBI
# Implement:
# • Show_Interest_Rate()

# Concept
# Learn abstract classes and methods.

from abc import ABC,abstractmethod

class Bank(ABC):
    def __init__(self, bank_name, branch):
        self.bank_name = bank_name
        self.branch = branch

    @abstractmethod
    def Show_Interest_Rate(self):
        pass

class SBI(Bank):
    def __init__(self, bank_name, branch):
        super().__init__(bank_name, branch)

    def Show_Interest_Rate(self):
        print("Interest Rate: 6.5%")

    def display(self):
        print("bank_name: ", self.bank_name)
        print("branch: ", self.branch)
        self.Show_Interest_Rate()

b1 = SBI("SBI", "Tirthpuri")
b1.display()
```

```python
# 20. University System

# Abstract Class
# University

# Attributes:
# • University_Name
# • City

# Abstract Method:
# • Show_University_Details()

# Child Class
# EngineeringCollege

# Attribute:
# • College_Name

# Implement abstract method and display all details.

# Concept
# Combine Abstraction + Inheritance + super().

from abc import ABC,abstractmethod
class University(ABC):
    def __init__(self, University_Name, City):
        self.University_Name = University_Name
        self.City = City

    @abstractmethod
    def Show_University_Details(self):
        pass

class EngineeringCollege(University):
    def __init__(self, University_Name, City, College_Name):
        super().__init__(University_Name, City)
        self.College_Name = College_Name

    def Show_University_Details(self):
        print("University_Name: ",self.University_Name)
        print("City: ",self.City)
        print("College_Name: ",self.College_Name)

u1 = EngineeringCollege("BAMU", "Chh.Sambhajinagar", "Deogiri College")
u1.Show_University_Details()
```
