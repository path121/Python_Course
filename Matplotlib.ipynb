```python
import matplotlib.pyplot as plt
```

```python
X = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
Y = [10,20,50,40,60,50]

plt.plot(X,Y)

# graph title
plt.title("Bakery Sales This Week!")
plt.xlabel("Day's")
plt.ylabel('sales Per Day')



plt.show()
```

```python
plt.scatter(X,Y)
plt.bar(X,Y)
plt.plot(X,Y,
        marker = "*",
        markerfacecolor =  "red",
        markeredgecolor = "red",
        markersize = 20,
        linewidth = 4,
        linestyle = ":",
        alpha = 0.6
        
        )
plt.title("Data")

plt.xlabel("Week Days")
plt.ylabel('Sale Per Day')

plt.grid()
plt.show()
```
