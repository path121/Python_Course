```python
# 8. Read a file line by line and print line numbers.
# Output
# 1 Python
# 2 Java
# 3 C++

file = open("student.txt", "r")

for line_no, line in enumerate(file, start=1):
    print(line_no, line.strip())

file.close()
```

```python
# 10. Find how many times a specific word appears in a file.
# Example

# Python Java Python C++ Python
# Find:
# Python
# Output:
# 3

# Create file
with open("destination.txt", "w") as file:
    file.write("Python Java Python C++ Python")

# Word to find
word = input("Find: ")

# Read file and count word
with open("destination.txt", "r") as file:
    data = file.read()

count = data.split().count(word)

print("Output:", count)
```
