## concatenate

#### syntax
- **.concatenate( (array1,array2), axis = 0)**
- **axis=0 :** means row-wise stack > (Vertical Stacking)
- **axis=1 :** means column-wise stack > (Horizontal stacking)

```python
import numpy as np
```

```python
arr1 = np.array([1,2,3,])
arr2 = np.array([4,5,6,])

new_arr = np.concatenate((arr1,arr2), axis = 0)
```

```python
print(new_arr)
```

- ***Concatenate can merge the array's***
