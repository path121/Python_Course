# Broadcasting
- It is a numpy way where performe operations on different-different shape array without using loops

---

**Without Broadcasting:**
 - using loop

```python
prices = [100,200,300]

discount = 10 #10% discount

final_prices = []

for price in prices:
    final_price = price - (price * discount) / 100
    final_prices.append(final_price)

print(final_prices)
```

---

**Using Broacasting :**
- Without using loop



```python
import numpy as np

prices = np.array([100,200,300])

discount = 10

final_prices = prices - (prices * discount) / 100

print(final_prices)
```
