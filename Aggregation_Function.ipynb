# Aggregation functions
- **sum()** : Used To Calculate Sum Of All Values
- **mean()** : Usd To Calculate Average Of All Value
- **min()** : Used To Get Minimun Value from Data
- **max()** : Used To Get Maximum Value from Data
- **std()** : Calculate Standard Deviation
- **var()** : Used to finding variance

```python
import numpy as np

arr = np.array([10,20,30,40,50])

print(np.sum(arr))
print(np.mean(arr))
print(np.min(arr))
print(np.max(arr))
print(np.std(arr))
print(np.var(arr))
```

```python
print(arr.mean())
```
