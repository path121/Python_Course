```python
# 1. Student Name Update
# Question:
# Create a Student class with attributes name and age.
# Create an object and update the student's name from "Rahul" to "Amit".
# What you need to do:
# • Create a class.
# • Create an object.
# • Change the value of an attribute.


class Student:
    def __init__(self,name,age):
        self.std_name = name
        self.std_age = age


    def get_student_name(self):
        print("Name :",self.std_name)

    def get_student_age(self):
        print("Age :",self.std_age)

    def update_name(self,new_name):
        self.std_name = new_name
        print("Successfully std name updated! :", new_name)
    
    def show_std_details(self):
        print("-" * 30)
        print("All Details :")
        print("-" * 30)
        
        print("New Name :", self.std_name)
        print("Age", self.std_age)
        print("-" * 30)

        
std1 = Student("Amit",21)

std1.get_student_name()
print("=" * 50)

std1.get_student_age()
print("=" * 50)

std1.update_name("Rahul")
print("=" * 50)

std1.show_std_details()
```

```python
# 2. Car Speed Check
# Question:
# Create a Car class with attributes brand and speed.
# If speed is greater than 80, print:
# "Over Speed"
# Otherwise print:
# "Safe Driving"
# What you need to learn:
# • Access object attributes.
# # • Use if condition.


class Car:

    def __init__(self,brand,speed):
        self.car_brand = brand
        self.car_speed = speed

    def check_speed(self):
        if self.car_speed > 80:
            print("Over Speed")
        else :
            print("Safe Driving")

    def get_car_brand(self):
        print("Brand:",self.car_brand)

    def get_car_speed(self):
        print("Speed:",self.car_speed)
# ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
print("1st car detail : \n")

car1 = Car("Toyota", 90)
car1.check_speed()
car1.get_car_brand()
car1.get_car_speed()

print("=" * 50,"\n")

print("2nd car detail : \n")
car2 = Car("Honda", 60)
car2.check_speed()
car2.get_car_brand()
car2.get_car_speed()
```

```python
# 4. Product Stock Check
# Question:
# Create a Product class with attributes:
# product_name
# stock
# If stock is 0:
# Out of Stock
# Otherwise:
# Available

class Product:

    def __init__(self, product_name, stock):
        
        self.product_name = product_name
            
        if stock <= 0:
            print("Out Of Stock")
        else :
            self.stock = stock
            print("Available")


        
    def display(self):
        print("product_name: ",self.product_name)
        print("stock: ",self.stock)
        


p1 = Product("Water Bottel",1)
p1.display()
```

```python
# 5. Mobile Battery Status
# Question:
# Create a Mobile class with attributes:
# model
# battery
# If battery is below 20:
# Low Battery
# Else:
# Battery OK

class Mobile:
    def __init__(self, model, battry):
        self.mobel = model

        if battry < 20:
            print("Low battry")

        else :
            self.battry = battry
            print("Battry Ok")

    def display(self):
        print("Model: ",self.mobel)
        print("Battry: ", self.battry)

m1 = Mobile("Motorola", 20)
m1.display()
```

```python
# Medium Level
# 6. Bank Account Deposit
# Question:
# Create a BankAccount class with:
# name
# balance
# Add ₹5000 to the balance.
# If balance becomes more than ₹20000:
# Premium Customer
# Otherwise:

class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def get_balance(self):
        print("Current Balance: ",self.balance)
        
    def add_balance(self, add_balance):
        if add_balance <= 0:
            print("Insufficient Balance!!...")
        else :
            self.balance += add_balance
            print("Balance Successfully Deposit")


        if self.balance > 20000:
            print("You Are Our Premium Customer")


    def display(self):
        print("Name: ", self.name)
        print("Balance: ", self.balance)
        
b1 = BankAccount("Dhanraj",  5000)
b1.display()
b1.add_balance(50000)
```

```python
# 7. Student Result System
# Question:
# Create a Student class with:
# name
# marks
# Condition:
# marks >= 90 → Grade A
# marks >= 75 → Grade B
# marks >= 50 → Grade C
# otherwise Fail
# Concepts:
# • Multiple conditions (if-elif-else)

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

        if marks >= 90:
            print("Grade A")

        elif marks >= 75 :
            print("Grade B")

        elif marks >= 50 :
            print("Grade C")

        else :
            print("Otherwise Fail")

    def display(self):
        print(self.name)
        print(self.marks)

s1 = Student("Dhanraj", 90)
s1.display()
```

```python
# 8. Laptop Price Discount
# Question:
# Create a Laptop class with:
# brand
# price
# If price is greater than 50000:
# Apply 10% discount.
# Display updated price.

class Laptop:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

        if self.price > 50000:
            discount = self.price * 10 / 100
            self.price = self.price - discount
    def display(self):
        print("Brand: ",self.brand)
        print("Price With Discount: ", self.price)


l1 = Laptop("Asus",60000)
l1.display()
```
