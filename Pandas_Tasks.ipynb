```python
# Q6. Fill missing values using appropriate techniques:
# • Name → "Unknown"
# • City → "Not Available"
# • Department → "Not Assigned"

# Fill missing values in Name
df["Name"] = df["Name"].fillna("Unknown")
```

```python
# Fill missing values in City
df["City"] = df["City"].fillna("Not Available")
```

```python
# Fill missing values in Department

df["Department"] = df["Department"].fillna("Not Assigned")
```

```python
# Q8. Find duplicate Customer_ID values and remove duplicate records while keeping the first occurrence.

# Display duplicate Customer_ID records
df[df.duplicated(subset="Customer_ID")]

```

```python
# Remove duplicate Customer_ID records
df.drop_duplicates(subset="Customer_ID", keep="first", inplace=True)
df
```

```python
# Verify
print(df.duplicated(subset="Customer_ID").sum())
```

```python
# Q9. Remove leading and trailing spaces from the Name column.

df["Name"] = df["Name"].str.strip()
```

```python
# Q10. Standardize the Gender column into only:
# • Male
# • Female

df.replace({"Gender":{
    "male": "Male",
    "MALE": "Male",
    "M": "Male",
    "female": "Female",
    "FEMALE": "Female",
    "F": "Female"
}},inplace = True)
```

```python
# Q11. Standardize city names.
# Example:
# HYD
# hyd
# Hyderabad
# ↓
# Hyderabad
# Do the same for Bengaluru/Bangalore if needed.

df.replace({"City": {
    "HYD":"Hyderabad",
    "hyd":"Hyderabad",
    "Hyderabad":"Hyderabad"
}},inplace = True)
```

```python
df.replace({"City": {
    "Bengaluru":"Bengaluru",
    "Bangalore":"Bengaluru"
}},inplace = True)
```

```python
df
```

```python
# Q12. Remove:
# • ₹ symbol
# • commas (,)
# Convert Salary into a numeric datatype.

df["Salary"] = df["Salary"].str.replace("₹", "")
df["Salary"] = df["Salary"].str.replace(",", "")
df["Salary"] = pd.to_numeric(df["Salary"])
```

```python
df
```

```python
# Part 7: Data Type Conversion
# Q13. Convert:
# • Age
# • Salary
# • Experience
# • Performance_Score

# • Rating
# to proper numeric datatypes.

df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
```

```python
cols = ["Age", "Salary", "Experience", "Performance_Score", "Rating"]

df[cols] = df[cols].apply(pd.to_numeric)
```

```python
# Part 8: Invalid Values
# Q14. Find all rows where:
# • Age < 18
# • Age > 100
# Replace invalid ages with NaN.

df.loc[(df["Age"] < 18) | (df["Age"] > 100), "Age"] = pd.NA
```
