```python
# 1. Handle ValueError
# Problem
# Ask the user to enter a number. If the user enters text instead of a number, display:
# Please enter a valid number!
# Concepts
# • try
# • except
# • ValueError


try:
    num = int(input("Enter The Number: "))
except ValueError:
    print("Please enter a valid number!")

else :
    print(f"Your Entred Number Is :{num}")
```

```python
# 2. Handle ZeroDivisionError
# Problem
# Take two numbers from the user and divide them.
# If the denominator is 0, print:
# Cannot divide by zero!
# Concepts
# • ZeroDivisionError
# • Exception Handling

try:
    A = int(input("Enter First Number: "))
    B = int(input("Enter Second Number: "))

    result = A / B
   
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

```python
# 3. Multiple Exceptions
# Problem
# Take a number from the user and divide 100 by it.
# Handle:
# • ValueError
# • ZeroDivisionError
# Concepts
# • Multiple except blocks

try:
    a = int(input("Enter Number: "))

    result = 100 / a
    print("Result: ",result)

except ValueError:
    print("Please enter a valid number!")

except ZeroDivisionError:
    print("Cannot divide by zero!")
```

```python
# 4. Print Error Message

# Problem
# Convert "hello" into an integer.
# Display the actual error message generated by Python.
# Concepts
# • except as e

try:
    num = int("hello")
    print(num)

except ValueError as e:
    print("Error:", e)
```

```python
# 5. File Not Found
# Problem
# Open a file named:
# student.txt
# If the file does not exist, display:
# File not found!
# Concepts
# • FileNotFoundError

try:
    file = open("student.txt", "r")
    print(file.read())
    file.close()
    print("*" * 55)

except FileNotFoundError:
    print("File not found!")
    
else :
    print("File is Available")

```

```python
# 6. Use else Block
# Problem
# Take a number from the user.
# If no error occurs:
# Square of number is ...
# Use an else block.
# Concepts
# • try
# • except
# • else

try:

    A = int(input("Enter A Number: "))

    Result = A ** 2

except ValueError:
    print("Please enter a valid number!")

else:
    print("Square = ",Result)
    
```

```python
# 7. Use finally Block
# Problem

# Open a file and read its content.
# Regardless of success or failure, print:
# Program Finished
# Concepts
# • finally

try:

    with open("student.txt","r") as file:
        content  = file.read()
        print(content)
        print("*" * 55)

except FileNotFoundError:
    print("File not found!")

finally:
    print("Program Finished")
```

```python
##### 8. Catch Multiple Errors in One Line
# Problem
# Write a program that can generate:
# • ValueError
# • ZeroDivisionError
# Handle both in a single except block.
# Concepts
# except (ValueError, ZeroDivisionError):

try:

    A = int(input("Enter Number: "))

    result = 100 / A

except(ValueError, ZeroDivisionError):
    print("Check The Entered No & Non Zero Number!")

else:
    print("Result: ",result)
```
