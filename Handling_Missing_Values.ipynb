### np.isnan()
- Used to detect nan values in arrray
- It gives answer in boolean form
### *Syntax:
- **np,isnan(array_name)**

```python
import numpy as np 
arr = np.array([1,np.nan,3,np.nan,5,6])
print(np.isnan(arr))
```

# Function use to replace nan to numeric value
- **nan_to_num()** : tjis is function

### * syntax:
- **np.nan_to_num(array_name, nan = value)**

```python
cleaned_arr = np.nan_to_num(arr , nan = 0)
```

```python
print(np.isnan(arr))
print('\n')
print(np.isnan(cleaned_arr))
```

**All nan Values Are replaced with 0**

# Handling Infinite Values Using Function:
#### np.isinf()
- It used to dectect infinite values are present in array by boolean form

### *Syntax:

np.isinf(array_name)

```python
arr = np.array([1,2,np.inf, 4, -np.inf, 6])

print(np.isinf(arr))
```

### Replacing Infinite number into finite number

```python
cleanedarr = np.nan_to_num(arr, posinf = 1000, neginf = -1000)

print(cleanedarr)
```
